Now we'll try to plot the data in different ways.

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title="Median vs. Sample Size", figsize=(5,5))
```

Here we see that the most popular majors, orbit around 35,000K a year. There seems to be a negative correlation between most popular major and average income, at least after a certain point. Seems to be that popularity of major increases income until the 35-50K range, then it starts to decrease.

So it seems there's a type of major that doesn't attract people and doesn't pay, but there's another type that doesn't attract people and pays a lot. Maybe it's related to the difficulty of said majors as you'd expect the most profitable ones to be the most popular ones.

```python
majorcat_women = {}
majorcat_men = {}
for a in range(len(recent_grads)):
    major = recent_grads.iloc[a]["Major_category"]
    if major in majorcat_women:
        majorcat_women[major] += recent_grads.iloc[a]["Women"]
    else:
        majorcat_women[major] = recent_grads.iloc[a]["Women"]
    if major in majorcat_men:
        majorcat_men[major] += recent_grads.iloc[a]["Men"]
    else:
        majorcat_men[major] = recent_grads.iloc[a]["Men"]
        
```

```python
cat_men_women = pd.DataFrame([majorcat_women, majorcat_men]).T
```

```python
cat_men_women.columns = ["Women", "Men"]
```

```python
cat_men_women
```

```python
cat_men_women.plot.bar(figsize = (10,5), width=0.8,color=["Red", "Blue"], alpha=0.5, edgecolor=None)
```
